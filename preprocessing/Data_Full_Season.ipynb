```python
## Downloading libraries
from pybaseball import batting_stats
import pandas as pd
import numpy as np
from itertools import permutations
from random import sample
```

```python
# Load Cubs data
cubs_df = pd.read_csv("/Users/quinyuter/Desktop/Chicago_Cubs_2025_STATS.csv")

# Define corrected player names based on CSV
selected_players = {
    'Carson Kelly': 'C',
    'Michael Busch*': '1B',
    'Nico Hoerner': '2B',
    'Dansby Swanson': 'SS',
    'Jon Berti': '3B',
    'Ian Happ#': 'LF',
    'Pete Crow-Armstrong*': 'CF',
    'Kyle Tucker*': 'RF',
    'Seiya Suzuki': 'DH'
}

# Filter for selected players
starters_cubs = cubs_df[cubs_df['Player'].isin(selected_players.keys())].copy()
starters_cubs['Pos'] = starters_cubs['Player'].map(selected_players)

# Rename for consistency
starters_cubs.rename(columns={'Player': 'Name'}, inplace=True)

# Calculate singles
starters_cubs['1B'] = starters_cubs['H'] - starters_cubs['2B'] - starters_cubs['3B'] - starters_cubs['HR']

# Calculate outcome percentages
starters_cubs['SO%'] = starters_cubs['SO'] / starters_cubs['PA']
starters_cubs['BB%'] = starters_cubs['BB'] / starters_cubs['PA']
starters_cubs['HBP%'] = starters_cubs['HBP'] / starters_cubs['PA']
starters_cubs['1B%'] = starters_cubs['1B'] / starters_cubs['PA']
starters_cubs['2B%'] = starters_cubs['2B'] / starters_cubs['PA']
starters_cubs['3B%'] = starters_cubs['3B'] / starters_cubs['PA']
starters_cubs['HR%'] = starters_cubs['HR'] / starters_cubs['PA']
starters_cubs['SF%'] = starters_cubs['SF'] / starters_cubs['PA']
starters_cubs['SH%'] = starters_cubs['SH'] / starters_cubs['PA']
starters_cubs['GIDP%'] = starters_cubs['GIDP'] / starters_cubs['PA']

# Calculate Field Out %
starters_cubs['Field Out%'] = 1 - (
    starters_cubs['SO%'] + starters_cubs['BB%'] + starters_cubs['HBP%'] +
    starters_cubs['1B%'] + starters_cubs['2B%'] + starters_cubs['3B%'] +
    starters_cubs['HR%'] + starters_cubs['SF%'] + starters_cubs['SH%'] + starters_cubs['GIDP%']
)

# Select final columns
starters_cubs = starters_cubs[['Name', 'Pos', 'SO%', 'BB%', 'HBP%', '1B%', '2B%', '3B%', 'HR%', 'SF%', 'SH%', 'GIDP%', 'Field Out%']]

# Show result
starters_cubs
```
